```python
aes_sbox = [
    [int('63', 16), int('7c', 16), int('77', 16), int('7b', 16), int('f2', 16), int('6b', 16), int('6f', 16), int('c5', 16), int(
        '30', 16), int('01', 16), int('67', 16), int('2b', 16), int('fe', 16), int('d7', 16), int('ab', 16), int('76', 16)],
    [int('ca', 16), int('82', 16), int('c9', 16), int('7d', 16), int('fa', 16), int('59', 16), int('47', 16), int('f0', 16), int(
        'ad', 16), int('d4', 16), int('a2', 16), int('af', 16), int('9c', 16), int('a4', 16), int('72', 16), int('c0', 16)],
    [int('b7', 16), int('fd', 16), int('93', 16), int('26', 16), int('36', 16), int('3f', 16), int('f7', 16), int('cc', 16), int(
        '34', 16), int('a5', 16), int('e5', 16), int('f1', 16), int('71', 16), int('d8', 16), int('31', 16), int('15', 16)],
    [int('04', 16), int('c7', 16), int('23', 16), int('c3', 16), int('18', 16), int('96', 16), int('05', 16), int('9a', 16), int(
        '07', 16), int('12', 16), int('80', 16), int('e2', 16), int('eb', 16), int('27', 16), int('b2', 16), int('75', 16)],
    [int('09', 16), int('83', 16), int('2c', 16), int('1a', 16), int('1b', 16), int('6e', 16), int('5a', 16), int('a0', 16), int(
        '52', 16), int('3b', 16), int('d6', 16), int('b3', 16), int('29', 16), int('e3', 16), int('2f', 16), int('84', 16)],
    [int('53', 16), int('d1', 16), int('00', 16), int('ed', 16), int('20', 16), int('fc', 16), int('b1', 16), int('5b', 16), int(
        '6a', 16), int('cb', 16), int('be', 16), int('39', 16), int('4a', 16), int('4c', 16), int('58', 16), int('cf', 16)],
    [int('d0', 16), int('ef', 16), int('aa', 16), int('fb', 16), int('43', 16), int('4d', 16), int('33', 16), int('85', 16), int(
        '45', 16), int('f9', 16), int('02', 16), int('7f', 16), int('50', 16), int('3c', 16), int('9f', 16), int('a8', 16)],
    [int('51', 16), int('a3', 16), int('40', 16), int('8f', 16), int('92', 16), int('9d', 16), int('38', 16), int('f5', 16), int(
        'bc', 16), int('b6', 16), int('da', 16), int('21', 16), int('10', 16), int('ff', 16), int('f3', 16), int('d2', 16)],
    [int('cd', 16), int('0c', 16), int('13', 16), int('ec', 16), int('5f', 16), int('97', 16), int('44', 16), int('17', 16), int(
        'c4', 16), int('a7', 16), int('7e', 16), int('3d', 16), int('64', 16), int('5d', 16), int('19', 16), int('73', 16)],
    [int('60', 16), int('81', 16), int('4f', 16), int('dc', 16), int('22', 16), int('2a', 16), int('90', 16), int('88', 16), int(
        '46', 16), int('ee', 16), int('b8', 16), int('14', 16), int('de', 16), int('5e', 16), int('0b', 16), int('db', 16)],
    [int('e0', 16), int('32', 16), int('3a', 16), int('0a', 16), int('49', 16), int('06', 16), int('24', 16), int('5c', 16), int(
        'c2', 16), int('d3', 16), int('ac', 16), int('62', 16), int('91', 16), int('95', 16), int('e4', 16), int('79', 16)],
    [int('e7', 16), int('c8', 16), int('37', 16), int('6d', 16), int('8d', 16), int('d5', 16), int('4e', 16), int('a9', 16), int(
        '6c', 16), int('56', 16), int('f4', 16), int('ea', 16), int('65', 16), int('7a', 16), int('ae', 16), int('08', 16)],
    [int('ba', 16), int('78', 16), int('25', 16), int('2e', 16), int('1c', 16), int('a6', 16), int('b4', 16), int('c6', 16), int(
        'e8', 16), int('dd', 16), int('74', 16), int('1f', 16), int('4b', 16), int('bd', 16), int('8b', 16), int('8a', 16)],
    [int('70', 16), int('3e', 16), int('b5', 16), int('66', 16), int('48', 16), int('03', 16), int('f6', 16), int('0e', 16), int(
        '61', 16), int('35', 16), int('57', 16), int('b9', 16), int('86', 16), int('c1', 16), int('1d', 16), int('9e', 16)],
    [int('e1', 16), int('f8', 16), int('98', 16), int('11', 16), int('69', 16), int('d9', 16), int('8e', 16), int('94', 16), int(
        '9b', 16), int('1e', 16), int('87', 16), int('e9', 16), int('ce', 16), int('55', 16), int('28', 16), int('df', 16)],
    [int('8c', 16), int('a1', 16), int('89', 16), int('0d', 16), int('bf', 16), int('e6', 16), int('42', 16), int('68', 16), int(
        '41', 16), int('99', 16), int('2d', 16), int('0f', 16), int('b0', 16), int('54', 16), int('bb', 16), int('16', 16)]
]

reverse_aes_sbox = [
    [int('52', 16), int('09', 16), int('6a', 16), int('d5', 16), int('30', 16), int('36', 16), int('a5', 16), int('38', 16), int(
        'bf', 16), int('40', 16), int('a3', 16), int('9e', 16), int('81', 16), int('f3', 16), int('d7', 16), int('fb', 16)],
    [int('7c', 16), int('e3', 16), int('39', 16), int('82', 16), int('9b', 16), int('2f', 16), int('ff', 16), int('87', 16), int(
        '34', 16), int('8e', 16), int('43', 16), int('44', 16), int('c4', 16), int('de', 16), int('e9', 16), int('cb', 16)],
    [int('54', 16), int('7b', 16), int('94', 16), int('32', 16), int('a6', 16), int('c2', 16), int('23', 16), int('3d', 16), int(
        'ee', 16), int('4c', 16), int('95', 16), int('0b', 16), int('42', 16), int('fa', 16), int('c3', 16), int('4e', 16)],
    [int('08', 16), int('2e', 16), int('a1', 16), int('66', 16), int('28', 16), int('d9', 16), int('24', 16), int('b2', 16), int(
        '76', 16), int('5b', 16), int('a2', 16), int('49', 16), int('6d', 16), int('8b', 16), int('d1', 16), int('25', 16)],
    [int('72', 16), int('f8', 16), int('f6', 16), int('64', 16), int('86', 16), int('68', 16), int('98', 16), int('16', 16), int(
        'd4', 16), int('a4', 16), int('5c', 16), int('cc', 16), int('5d', 16), int('65', 16), int('b6', 16), int('92', 16)],
    [int('6c', 16), int('70', 16), int('48', 16), int('50', 16), int('fd', 16), int('ed', 16), int('b9', 16), int('da', 16), int(
        '5e', 16), int('15', 16), int('46', 16), int('57', 16), int('a7', 16), int('8d', 16), int('9d', 16), int('84', 16)],
    [int('90', 16), int('d8', 16), int('ab', 16), int('00', 16), int('8c', 16), int('bc', 16), int('d3', 16), int('0a', 16), int(
        'f7', 16), int('e4', 16), int('58', 16), int('05', 16), int('b8', 16), int('b3', 16), int('45', 16), int('06', 16)],
    [int('d0', 16), int('2c', 16), int('1e', 16), int('8f', 16), int('ca', 16), int('3f', 16), int('0f', 16), int('02', 16), int(
        'c1', 16), int('af', 16), int('bd', 16), int('03', 16), int('01', 16), int('13', 16), int('8a', 16), int('6b', 16)],
    [int('3a', 16), int('91', 16), int('11', 16), int('41', 16), int('4f', 16), int('67', 16), int('dc', 16), int('ea', 16), int(
        '97', 16), int('f2', 16), int('cf', 16), int('ce', 16), int('f0', 16), int('b4', 16), int('e6', 16), int('73', 16)],
    [int('96', 16), int('ac', 16), int('74', 16), int('22', 16), int('e7', 16), int('ad', 16), int('35', 16), int('85', 16), int(
        'e2', 16), int('f9', 16), int('37', 16), int('e8', 16), int('1c', 16), int('75', 16), int('df', 16), int('6e', 16)],
    [int('47', 16), int('f1', 16), int('1a', 16), int('71', 16), int('1d', 16), int('29', 16), int('c5', 16), int('89', 16), int(
        '6f', 16), int('b7', 16), int('62', 16), int('0e', 16), int('aa', 16), int('18', 16), int('be', 16), int('1b', 16)],
    [int('fc', 16), int('56', 16), int('3e', 16), int('4b', 16), int('c6', 16), int('d2', 16), int('79', 16), int('20', 16), int(
        '9a', 16), int('db', 16), int('c0', 16), int('fe', 16), int('78', 16), int('cd', 16), int('5a', 16), int('f4', 16)],
    [int('1f', 16), int('dd', 16), int('a8', 16), int('33', 16), int('88', 16), int('07', 16), int('c7', 16), int('31', 16), int(
        'b1', 16), int('12', 16), int('10', 16), int('59', 16), int('27', 16), int('80', 16), int('ec', 16), int('5f', 16)],
    [int('60', 16), int('51', 16), int('7f', 16), int('a9', 16), int('19', 16), int('b5', 16), int('4a', 16), int('0d', 16), int(
        '2d', 16), int('e5', 16), int('7a', 16), int('9f', 16), int('93', 16), int('c9', 16), int('9c', 16), int('ef', 16)],
    [int('a0', 16), int('e0', 16), int('3b', 16), int('4d', 16), int('ae', 16), int('2a', 16), int('f5', 16), int('b0', 16), int(
        'c8', 16), int('eb', 16), int('bb', 16), int('3c', 16), int('83', 16), int('53', 16), int('99', 16), int('61', 16)],
    [int('17', 16), int('2b', 16), int('04', 16), int('7e', 16), int('ba', 16), int('77', 16), int('d6', 16), int('26', 16), int(
        'e1', 16), int('69', 16), int('14', 16), int('63', 16), int('55', 16), int('21', 16), int('0c', 16), int('7d', 16)]
]


# Funcția break_in_grids_of_16 primește o secvență de octeți s și împarte acea secvență în grile de 16 octeți fiecare. 
# Fiecare grilă este reprezentată ca o listă bidimensională de dimensiune 4x4, unde fiecare element al grilei este un octet din secvența inițială. 
# Funcția returnează o listă cu toate grilele obținute.
def break_in_grids_of_16(s):
    all = []
    for i in range(len(s)//16):
        b = s[i*16: i*16 + 16]
        grid = [[], [], [], []]
        for i in range(4):
            for j in range(4):
                grid[i].append(b[i + j*4])
        all.append(grid)
    return all


# Funcția lookup primește un octet byte și utilizează operații pe biți pentru a extrage valorile x și y din octet.
def lookup(byte):
    x = byte >> 4
    y = byte & 15
    return aes_sbox[x][y]


# Funcția reverse_lookup primește un octet byte și utilizează operații pe biți pentru a extrage valorile x și y din octet. 
# Aceste valori sunt apoi folosite ca indici pentru a accesa tabela inversă a AES S-Box, numită reverse_aes_sbox, și a returna valoarea corespunzătoare din tabelă.
def reverse_lookup(byte):
    x = byte >> 4
    y = byte & 15
    return reverse_aes_sbox[x][y]


# Funcția realizează o rotație la stânga a elementelor din row cu n poziții și returnează rezultatul.
def rotate_row_left(row, n=1):
    return row[n:] + row[:n]


# Funcția expand_key primește o cheie key și numărul de runde rounds. 
# Aceasta extinde cheia folosind algoritmul AES Key Expansion și returnează rezultatul sub forma unei matrice de bytes.
# Cheia este împărțită în blocuri de 16 bytes folosind funcția break_in_grids_of_16. 
# Se selectează primul bloc de cheie pentru a fi procesat.
# Returnează o matrice de bytes reprezentând cheia extinsă.
def expand_key(key, rounds):

    rcon = [[1, 0, 0, 0]]

    for _ in range(1, rounds):
        rcon.append([rcon[-1][0]*2, 0, 0, 0])
        if rcon[-1][0] > 0x80:
            rcon[-1][0] ^= 0x11b

    key_grid = break_in_grids_of_16(key)[0]

    for round in range(rounds):
        last_column = [row[-1] for row in key_grid]
        last_column_rotate_step = rotate_row_left(last_column)
        last_column_sbox_step = [lookup(b) for b in last_column_rotate_step]
        last_column_rcon_step = [last_column_sbox_step[i]
                                 ^ rcon[round][i] for i in range(len(last_column_rotate_step))]

        for r in range(4):
            key_grid[r] += bytes([last_column_rcon_step[r]
                                  ^ key_grid[r][round*4]])

        # Three more columns to go
        for i in range(len(key_grid)):
            for j in range(1, 4):
                key_grid[i] += bytes([key_grid[i][round*4+j]
                                      ^ key_grid[i][round*4+j+3]])

    return key_grid


# Funcția multiply_by_2 primește o valoare v și o înmulțește cu 2 în cadrul unui câmp finit de 8 biți. Ea returnează rezultatul înmulțirii.
def multiply_by_2(v):
    s = v << 1
    s &= 0xff
    if (v & 128) != 0:
        s = s ^ 0x1b
    return s


# Funcția multiply_by_3 primește o valoare v și realizează înmulțirea cu 3 în cadrul unui câmp finit de 8 biți. 
# Pentru a realiza această operație, funcția utilizează operația de înmulțire cu 2 (multiply_by_2) și efectuează un XOR între rezultatul înmulțirii cu 2 și valoarea v inițială.
def multiply_by_3(v):
    return multiply_by_2(v) ^ v


# Funcția mix_columns primește o grilă de 4x4 (grid) și aplică operația de amestecare a coloanelor în algoritmul AES.
def mix_columns(grid):
    new_grid = [[], [], [], []]
    for i in range(4):
        col = [grid[j][i] for j in range(4)]
        col = mix_column(col)
        for i in range(4):
            new_grid[i].append(col[i])
    return new_grid


# Funcția mix_column primește o coloană (column) reprezentată ca o listă de 4 elemente și aplică operația de amestecare a coloanei conform algoritmului AES. 
def mix_column(column):
    r = [
        multiply_by_2(column[0]) ^ multiply_by_3(
            column[1]) ^ column[2] ^ column[3],
        multiply_by_2(column[1]) ^ multiply_by_3(
            column[2]) ^ column[3] ^ column[0],
        multiply_by_2(column[2]) ^ multiply_by_3(
            column[3]) ^ column[0] ^ column[1],
        multiply_by_2(column[3]) ^ multiply_by_3(
            column[0]) ^ column[1] ^ column[2],
    ]
    return r


# Funcția add_sub_key primește o grilă de bloc (block_grid) și o grilă de cheie (key_grid) și aplică operația de adunare XOR între fiecare element corespunzător din cele două grile.
def add_sub_key(block_grid, key_grid):
    r = []
    for i in range(4):
        r.append([])
        for j in range(4):
            r[-1].append(block_grid[i][j] ^ key_grid[i][j])
    return r


# Funcția extract_key_for_round primește o cheie extinsă (expanded_key) și un număr de rundă (round) și extrage subcheia corespunzătoare acelei runde. 
def extract_key_for_round(expanded_key, round):
  return [row[round*4: round*4 + 4] for row in expanded_key]


# Funcția enc primește o cheie și datele de intrare și aplică algoritmul AES pentru a cripta datele utilizând cheia specificată. 
# Ea împarte datele în blocuri de 16 octeți, extinde cheia pentru runde, și aplică pași specifici algoritmului AES pentru fiecare bloc de date în fiecare rundă.
# La final, rezultatul criptării este returnat.
def enc(key, data):
    # Impartim datele si le spargem in blocuri de 16
    pad = bytes(16 - len(data) % 16)

    if len(pad) != 16:
        data += pad
    grids = break_in_grids_of_16(data)

    # Facem expand la cheie
    expanded_key = expand_key(key, 11)

    # Aplicam cheia originala pe blocuri
    # De acum folosim integer
    temp_grids = []
    round_key = extract_key_for_round(expanded_key, 0)

    for grid in grids:
        temp_grids.append(add_sub_key(grid, round_key))

    grids = temp_grids

    # Partea principala din algoritm
    for round in range(1, 10):
        temp_grids = []

        for grid in grids:
            sub_bytes_step = [[lookup(val) for val in row] for row in grid]
            shift_rows_step = [rotate_row_left(
                sub_bytes_step[i], i) for i in range(4)]
            mix_column_step = mix_columns(shift_rows_step)
            round_key = extract_key_for_round(expanded_key, round)
            add_sub_key_step = add_sub_key(mix_column_step, round_key)
            temp_grids.append(add_sub_key_step)

        grids = temp_grids

    # Runda finala fara sa mixam coloanele
    temp_grids = []
    round_key = extract_key_for_round(expanded_key, 10)

    for grid in grids:
        sub_bytes_step = [[lookup(val) for val in row] for row in grid]
        shift_rows_step = [rotate_row_left(
            sub_bytes_step[i], i) for i in range(4)]
        add_sub_key_step = add_sub_key(shift_rows_step, round_key)
        temp_grids.append(add_sub_key_step)

    grids = temp_grids

    # recream data intr-un singur stream inainte de return
    int_stream = []

    for grid in grids:
        for column in range(4):
            for row in range(4):
                int_stream.append(grid[row][column])

    return bytes(int_stream)


# Funcția dec primește o cheie și datele criptate și efectuează decriptarea utilizând algoritmul AES. 
# Ea inversează pașii de criptare în ordine inversă, aplicând operațiile de decriptare corespunzătoare pentru fiecare rundă. 
# La final, rezultatul decriptării este returnat.
def dec(key, data):

    grids = break_in_grids_of_16(data)
    expanded_key = expand_key(key, 11)
    temp_grids = []
    round_key = extract_key_for_round(expanded_key, 10)

    # Facem undo 
    temp_grids = []

    for grid in grids:

        add_sub_key_step = add_sub_key(grid, round_key)
        shift_rows_step = [rotate_row_left(
            add_sub_key_step[i], -1 * i) for i in range(4)]
        sub_bytes_step = [[reverse_lookup(val) for val in row]
                          for row in shift_rows_step]
        temp_grids.append(sub_bytes_step)

    grids = temp_grids

    for round in range(9, 0, -1):
        temp_grids = []

        for grid in grids:
            round_key = extract_key_for_round(expanded_key, round)
            add_sub_key_step = add_sub_key(grid, round_key)

            # Facem mix la coloane de 3 pentru ca este egal cu folosirea matricii inverse
            mix_column_step = mix_columns(add_sub_key_step)
            mix_column_step = mix_columns(mix_column_step)
            mix_column_step = mix_columns(mix_column_step)
            shift_rows_step = [rotate_row_left(
                mix_column_step[i], -1 * i) for i in range(4)]
            sub_bytes_step = [
                [reverse_lookup(val) for val in row] for row in shift_rows_step]
            temp_grids.append(sub_bytes_step)

        grids = temp_grids
        temp_grids = []

    # Reverse la prima si adaugam cheia
    round_key = extract_key_for_round(expanded_key, 0)

    for grid in grids:
        temp_grids.append(add_sub_key(grid, round_key))

    grids = temp_grids

    # Transformam gridurile inapooi in bytes
    int_stream = []
    for grid in grids:
        for column in range(4):
            for row in range(4):
                int_stream.append(grid[row][column])

    return bytes(int_stream)

# Datele de intrare si de iesire
key = b"predefined_value"
message = b"abcdefghijklmnop"

encrypted_msg = enc(key,message)
print("Encrypted message is: {}".format(encrypted_msg))

print("Decrypted message is: {}".format(dec(key,encrypted_msg)))
```

## Algoritmul RSA

```python
import random

# Generează un număr prim aleatoriu.
def generate_prime_number():
    while True:
        num = random.randint(2, 100)
        if is_prime(num):
            return num

# Verifică dacă un număr este prim.
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True
# Calculează cel mai mare divizor comun dintre două numere.

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

# Funcția multiplicative_inverse(e, phi) calculează inversul multiplicativ al unui număr e modulo phi. 
# Acesta este utilizat în criptografia asimetrică pentru a calcula cheia privată pe baza cheii publice și a valorii phi. 
# Algoritmul extins al lui Euclid este folosit pentru a găsi coeficienții necesari, iar rezultatul este returnat ca inversul multiplicativ.
def multiplicative_inverse(e, phi):
    d = 0
    x1 = 0
    x2 = 1
    y1 = 1
    temp_phi = phi
    
    while e > 0:
        temp1 = temp_phi // e
        temp2 = temp_phi - temp1 * e
        temp_phi = e
        e = temp2
        
        x = x2 - temp1 * x1
        y = d - temp1 * y1
        
        x2 = x1
        x1 = x
        d = y1
        y1 = y

    if temp_phi == 1:
        return d + phi

# Generează o pereche de chei RSA (publică și privată).
def generate_keypair():
    p = generate_prime_number()
    q = generate_prime_number()
    n = p * q
    phi = (p - 1) * (q - 1)

    # Alegem un exponent public e
    # care să fie relativ prim cu phi
    while True:
        e = random.randint(2, phi)
        if gcd(e, phi) == 1:
            break

    # Calculăm exponentul privat d
    d = multiplicative_inverse(e, phi)

    return ((e, n), (d, n))

# Criptează un mesaj folosind cheia publică RSA.
def encrypt(message, public_key):
    e, n = public_key
    encrypted_message = [pow(ord(char), e, n) for char in message]
    return encrypted_message

# Decriptează un mesaj criptat folosind cheia privată RSA.
def decrypt(encrypted_message, private_key):
    d, n = private_key
    decrypted_message = [chr(pow(char, d, n)) for char in encrypted_message]
    return "".join(decrypted_message)


# Generăm o pereche de chei RSA
public_key, private_key = generate_keypair()

# Mesajul de test
message = "Hello man"

# Criptăm mesajul folosind cheia publică
encrypted_message = encrypt(message, public_key)
print("Encrypted message:", encrypted_message)

# Decriptăm mesajul folosind cheia privată
decrypted_message = decrypt(encrypted_message, private_key)
print("decrypted message:", decrypted_message)
```

## Algoritmul RC5

```python
import base64
import os
from io import BytesIO


class RC5(object):
    def __init__(self, key):
        self.mode = 'CBC'  # Modul de operare: "ECB" sau "CBC"
        self.blocksize = 32  # Dimensiunea blocului în biți
        self.rounds = 12  # Numărul de runde
        self.iv = os.urandom(self.blocksize // 8)  # Inițializarea vectorului de inițializare (IV)
        self._key = key.encode('utf-8')  # Cheia de criptare, convertită în format bytes

    @staticmethod
    def _rotate_left(val, r_bits, max_bits):
      # Funcție pentru rotirea la stânga a unei valori cu un anumit număr de biți
        v1 = (val << r_bits % max_bits) & (2 ** max_bits - 1)  # Rotirea la stânga cu r_bits modulo max_bits
        v2 = ((val & (2 ** max_bits - 1)) >> (max_bits - (r_bits % max_bits)))   # Se selectează cei r_bits % max_bits biți din dreapta și se deplasează la poziția corectă în valoare
        return v1 | v2  # Se combină cele două rezultate obținute pentru a obține rezultatul final al rotației la stânga

    @staticmethod
    def _rotate_right(val, r_bits, max_bits):
      # Funcție pentru rotirea la dreapta a unei valori cu un anumit număr de biți
        v1 = ((val & (2 ** max_bits - 1)) >> r_bits % max_bits)
        v2 = (val << (max_bits - (r_bits % max_bits)) & (2 ** max_bits - 1))

        return v1 | v2

    @staticmethod
    def _expand_key(key, wordsize, rounds):
        # Funcție pentru alinierea cheii și împărțirea acesteia în cuvinte
        def _align_key(key, align_val):
          # Se adaugă octeți cu valoarea 0 până când lungimea cheii este aliniată la dimensiunea blocului
            while len(key) % (align_val):
                key += b'\x00'  # Add 0 bytes until the _key length is aligned to the block size

            L = []
            for i in range(0, len(key), align_val):
              # Se împarte cheia în cuvinte de dimensiunea specificată și se stochează într-o listă
                L.append(int.from_bytes(key[i:i + align_val], byteorder='little'))

            return L

        # Această funcție _const primește ca parametru w (dimensiunea unui cuvânt) și returnează o pereche de valori. 
        # Aceste valori reprezintă constantele P și Q utilizate în algoritmul RC5 pentru extinderea cheii.
        def _const(w):
            if w == 16:
                return (0xB7E1, 0x9E37)  # Returns the value of P and Q
            elif w == 32:
                return (0xB7E15163, 0x9E3779B9)
            elif w == 64:
                return (0xB7E151628AED2A6B, 0x9E3779B97F4A7C15)

        # Această funcție _extend_key primește două argumente: w (dimensiunea unui cuvânt) și r (numărul de runde). 
        # Ea returnează o listă S care reprezintă programul pseudo-aleator utilizat în procesul de extindere a cheii în algoritmul RC5.
        def _extend_key(w, r):
            P, Q = _const(w)
            S = [P]
            t = 2 * (r + 1)
            for i in range(1, t):
                S.append((S[i - 1] + Q) % 2 ** w)

            return S

        # Funcția _mix este responsabilă pentru amestecarea listelor L și S pentru a genera o listă finală de cuvinte. 
        # Algoritmul utilizează o serie de rotații și operații aritmetice pentru a produce noile valori în lista S. 
        # Valorile A și B sunt inițializate cu 0, iar indicii i și j sunt inițializați cu 0.
        def _mix(L, S, r, w, c):
            t = 2 * (r + 1)
            m = max(c, t)
            A = B = i = j = 0

            for k in range(3 * m):
                A = S[i] = RC5._rotate_left(S[i] + A + B, 3, w)
                B = L[j] = RC5._rotate_left(L[j] + A + B, A + B, w)

                i = (i + 1) % t
                j = (j + 1) % c

            return S

        aligned = _align_key(key, wordsize // 8)
        extended = _extend_key(wordsize, rounds)

        S = _mix(aligned, extended, rounds, wordsize, len(aligned))

        return S

    # Funcția _encrypt_block implementează algoritmul de criptare pentru un bloc de date. 
    # Blocul de date este împărțit în două părți: A și B, fiecare având o dimensiune de blocksize // 2 biți.
    # Apoi, prin intermediul buclei for, se parcurg rounds iterații. 
    # În fiecare iterație, valorile A și B sunt actualizate utilizând operații de rotație, XOR și adunare modulo. 
    # Se folosește funcția _rotate_left pentru a roti valorile A și B în stânga cu un număr specific de biți (B și respectiv A în rotație). 
    # Valorile rezultate sunt adunate modulo mod cu valorile corespunzătoare din cheia extinsă (expanded_key[2 * i] și expanded_key[2 * i + 1]).
    @staticmethod
    def _encrypt_block(data, expanded_key, blocksize, rounds):
        w = blocksize // 2
        b = blocksize // 8
        mod = 2 ** w

        A = int.from_bytes(data[:b // 2], byteorder='little')
        B = int.from_bytes(data[b // 2:], byteorder='little')

        A = (A + expanded_key[0]) % mod
        B = (B + expanded_key[1]) % mod

        for i in range(1, rounds + 1):
            A = (RC5._rotate_left((A ^ B), B, w) + expanded_key[2 * i]) % mod
            B = (RC5._rotate_left((A ^ B), A, w) + expanded_key[2 * i + 1]) % mod

        res = A.to_bytes(b // 2, byteorder='little') + B.to_bytes(b // 2, byteorder='little')
        return res

    # În această funcție, se parcurge bucla for în sens invers, de la rounds la 1. 
    # În fiecare iterație, valorile A și B sunt actualizate utilizând operații de rotație, XOR și scădere. 
    # Se folosește funcția _rotate_right pentru a roti valorile B și A în dreapta cu un număr specific de biți (A și respectiv B în rotație). 
    # Valorile rezultate sunt apoi XOR-ate cu valorile corespunzătoare din cheia extinsă (expanded_key[2 * i] și expanded_key[2 * i + 1]).
    @staticmethod
    def _decrypt_block(data, expanded_key, blocksize, rounds):
        w = blocksize // 2
        b = blocksize // 8
        mod = 2 ** w

        A = int.from_bytes(data[:b // 2], byteorder='little')
        B = int.from_bytes(data[b // 2:], byteorder='little')

        for i in range(rounds, 0, -1):
            B = RC5._rotate_right(B - expanded_key[2 * i + 1], A, w) ^ A
            A = RC5._rotate_right((A - expanded_key[2 * i]), B, w) ^ B

        B = (B - expanded_key[1]) % mod
        A = (A - expanded_key[0]) % mod

        res = A.to_bytes(b // 2, byteorder='little') + B.to_bytes(b // 2, byteorder='little')
        return res

    # Functia encrypt_file primește trei argumente: infile (fișierul de intrare), outfile (fișierul de ieșire) și self (instanța clasei care apelează metoda).
    # Dacă modul de criptare este CBC, se inițializează variabila last_v cu vectorul de inițializare (iv).
    # De asemenea, se scrie vectorul de inițializare la începutul fișierului de ieșire (outfile.write(last_v)).
    # În fiecare iterație, se completează blocul citit anterior cu octeți de 0 până la dimensiunea dorită b utilizând metoda ljust. 
    # Apoi, dacă modul de criptare este CBC, se efectuează operația de XOR între last_v și blocul curent.
    # La finalul fiecărei iterații, se citește următorul bloc de dimensiunea b din fișierul de intrare utilizând infile.read(b). 
    # Acest proces continuă până când nu mai sunt blocuri disponibile în fișierul de intrare.
    def encrypt_file(self, infile, outfile):
        w = self.blocksize // 2
        b = self.blocksize // 8

        if self.mode == 'CBC':
            last_v = self.iv
            # set iv in the beginning of outfile
            outfile.write(last_v)

        expanded_key = RC5._expand_key(self._key, w, self.rounds)

        chunk = infile.read(b)

        while chunk:
            chunk = chunk.ljust(b, b'\x00')  # padding with 0 bytes if not large enough
            if self.mode == 'CBC':
                chunk = bytes([a ^ b for a, b in zip(last_v, chunk)])

            encrypted_chunk = RC5._encrypt_block(chunk, expanded_key,
                                                 self.blocksize,
                                                 self.rounds)
            outfile.write(encrypted_chunk)
            last_v = encrypted_chunk

            chunk = infile.read(b)  # Read in blocksize number of bytes

    # În această metodă, se inițializează variabilele w și b pentru a reprezenta dimensiunea blocului în biți și octeți.
    # În fiecare iterație, se decriptează blocul de date apelând funcția _decrypt_block cu parametrii corespunzători: 
    # chunk (blocul curent), expanded_key (cheia extinsă), self.blocksize (dimensiunea blocului) și self.rounds (numărul de runde).
    # Dacă nu mai există blocuri disponibile în fișierul de intrare (if not chunk), se elimină octeții de padding 0 din blocul decriptat utilizând decrypted_chunk.rstrip(b'\x00')
    def decrypt_file(self, infile, outfile):
        w = self.blocksize // 2
        b = self.blocksize // 8
        if self.mode == 'CBC':
            last_v = outfile.read(b)

        expanded_key = RC5._expand_key(self._key, w,
                                       self.rounds)

        chunk = infile.read(b)

        while chunk:
            decrypted_chunk = RC5._decrypt_block(chunk, expanded_key,
                                                 self.blocksize,
                                                 self.rounds)
            if self.mode == 'CBC':
                decrypted_chunk = bytes([a ^ b
                                         for a, b in zip(last_v,
                                                         decrypted_chunk)])
                last_v = chunk
            chunk = infile.read(b)  # Read in blocksize number of bytes
            if not chunk:
                decrypted_chunk = decrypted_chunk.rstrip(b'\x00')

            outfile.write(decrypted_chunk)

    # Funcția encrypt_str primește un șir de caractere de intrare și îl criptează utilizând metoda encrypt_file. 
    # Rezultatul criptării este returnat sub formă de șir de caractere codificat în formatul URL-safe Base64.
    def encrypt_str(self, input_str):
        str_in = BytesIO()
        str_in.write(input_str.encode('utf-8'))
        str_in.seek(0)
        str_out = BytesIO()

        self.encrypt_file(str_in, str_out)

        return base64.urlsafe_b64encode(str_out.getvalue()).decode("utf-8")

    # Funcția decrypt_str primește un șir de caractere criptat în formatul URL-safe Base64 și îl decriptează utilizând metoda decrypt_file. 
    # Rezultatul decriptării este returnat sub formă de șir de caractere decodificat utilizând encodarea UTF-8.
    def decrypt_str(self, input_enc_str):
        enc_bytes = base64.urlsafe_b64decode(input_enc_str)

        byte_in = BytesIO()
        byte_in.write(enc_bytes)
        byte_in.seek(0)
        byte_out = BytesIO()

        self.decrypt_file(byte_in, byte_out)

        return byte_out.getvalue().decode('utf-8')

# Criptează un șir de caractere inițial utilizând criptarea RC5 în modul CBC (Cipher Block Chaining).
# Afișează textul criptat rezultat și verifică lungimea acestuia.
# Decriptează textul criptat și afișează rezultatul decriptării.
# Verifică dacă textul decriptat este identic cu șirul de caractere inițial.
# Repetă același proces pentru criptarea în modul ECB (Electronic Codebook) pentru a efectua un test suplimentar.
if __name__ == '__main__':
    import time

    test_origin_char = '{}-{}'.format(99999999, int(time.time()))
    pwd = 'pig'

    # test encrypt string CBC mode
    cryptor = RC5(pwd)
    cryptor.mode = "CBC"

    print('text', test_origin_char.encode('utf-8'))
    enc_str = cryptor.encrypt_str(test_origin_char)
    print("encrypted_str", enc_str, len(enc_str))
    dec_str = cryptor.decrypt_str(enc_str)
    print("decrypted_str", dec_str)
    assert dec_str == test_origin_char

    # test encrypt string ECB mode
    cryptor = RC5(pwd)
    cryptor.mode = "ECB"

    print('text', test_origin_char.encode('utf-8'))
    enc_str = cryptor.encrypt_str(test_origin_char)
    print("encrypted_str", enc_str, len(enc_str))
    dec_str = cryptor.decrypt_str(enc_str)
    print("decrypted_str", dec_str)
    assert dec_str == test_origin_char
```
